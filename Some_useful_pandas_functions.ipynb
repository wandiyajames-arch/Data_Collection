### <font color = '#DEB887'> Pandas - Data Cleaning

```python
# import data.csv
df_csv2 = pd.read_csv('data.csv')
df_csv2.head()
```

```python
# Check the existence of null values number of variables that have missing values 
#df_csv2.isnull().sum() or  use .any()
#df_csv2.isnull().any().sum() or isna()
df_csv2.isna().any().sum()
```

```python
# Delete null values using dropna 
df_csv2.dropna(inplace=True)
df_csv2.head()
```

```python
# Replace null values by the mean
# import data.csv
df_csv2 = pd.read_csv('data.csv')
df_csv2.head()
```

```python
# Replace null values by the median 
# import data.csv
for col in df_csv2.columns[:2]:
    df_csv2.fillna(df_csv2[col].median(), inplace=True)
df_csv2.isna().sum()
```

```python
df_csv2.calories.mode()
```

```python
# Replace null values by the mode 
# import data.csv
for col in df_csv2.columns[:2]:
    df_csv2.fillna(df_csv2[col].mode()[0], inplace=True)
df_csv2.isna().sum()
```

```python
# Check the types of the different variables 
df_csv2.dtypes
```

```python
# import data1.csv
df_csv1 = pd.read_csv('data1.csv')
df_csv1
```

```python
# Convert calories in integer using astype 
df_csv1.calories = df_csv1.calories.astype('int64')
df_csv1.dtypes
```
