```python
# Import Pandas
import pandas as pd
```

```python
# Scrape the data tables from 2000 to 2024 using pandas.read_html and clean the d
url = 'https://www.basketball-reference.com/leagues/NBA_2010_per_game.html'
data = pd.read_html(url, header= 0)
len(data)
```

```python
# Display df.
#use for loop to collect data from 2000 to 2024 using pandas.read-html
df = pd.DataFrame()
for year in range(2000 , 2025):
  url_page = f'https://www.basketball-reference.com/leagues/NBA_{year}_per_game.html'
  data = pd.read_html(url_page)
  df_page = data[0]
  df_page['year'] = year
  df = pd.concat([df, df_page], axis = 0).reset_index(drop = True)
```
