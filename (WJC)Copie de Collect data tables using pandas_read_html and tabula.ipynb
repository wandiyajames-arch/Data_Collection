### <font color = 'gold'> Collect data over one page

```python
# url
url = 'https://www.basketball-reference.com/teams/'
```

```python
# Import pandas
import pandas as pd 
```

```python
# Collect all tables that exist on the wep page using pd.read_html
data = pd.read_html(url, header = 0)
```

```python
# Type of data
type(data)
```

```python
# Check the length of data
len(data)
```

```python
# First dataframe
df1 = data[0]
df1
```

```python
# the second dataframe
df2 = data[1]
df2
```

```python
# Check the dimension of df1 and df2
df1.shape , df2.shape
```

```python
# Check the existence of duplicated values on df1 and df2
print(df1.duplicated().sum())
print(df2.duplicated().sum())
```

```python
# Check the types of df1 and df2 variables
df1.dtypes, df2.dtypes
```

```python
#Check the existence of null values on df1 and df2
df1.fillna(df1.Div.mean(), inplace = True)
print(df1.isnull().any().sum())
```

```python
# Replace null values by the mean
df2.fillna(df2.Div.mean(), inplace = True)
print(df2.isnull().any().sum())
```

```python
# Concat df1 and df2
df = pd.concat([df1, df2], axis = 0)
df.tail()
```

### <font color = 'gold'> Collect data over multiple pages

```python
# In this web page, we have data of american football players
url = 'https://www.nfl.com/stats/player-stats/category/passing/2010/post/all/passingyards/DESC'
data = pd.read_html(url, header = 0)
len(data)
```

```python
# empty DataFrame : just do as df = pd.Dataframe() and df to print an empty DataFrame
#Use a for loop to  collect data from 1970  2022
df = pd.DataFrame()
for year in range(1970, 2023):
    url_page = f'https://www.nfl.com/stats/player-stats/category/passing/{year}/post/all/passingyards/DESC'
    data = pd.read_html(url_page)
    df_page = data[0]
    df_page['year'] = year
    df = pd.concat([df, df_page], axis = 0).reset_index(drop = True)
```
